```python
import numpy as np
import pandas as pd
import uuid
import random
from datetime import timedelta

open_orders = dict()
event_types = ['order_created', 'order_fulfilled']
periods = [ts for ts in pd.date_range(start='1/1/2008', end='31/12/2018') if random.random()>0.8]

def new_random_order():
    return {
        'order_id': uuid.uuid4(),
        'amount': random.random() * 1000,
        'created_at': random.choice(periods)
    } 

def new_random_event():
    event_type = random.choice(event_types)
    order_info = new_random_order()
    if event_type=='order_created':
        open_orders[order_info['order_id']] = order_info
    else:
        if len(open_orders):
            order_id = random.choice(list(open_orders.keys()))
            order_info = open_orders[order_id]
            order_info['created_at'] = order_info['created_at'] + timedelta(random.randrange(0,180))
        open_orders.pop(order_info['order_id'], None)
        
    return {
        "event_id": uuid.uuid4(),
        "event_type": event_type,
        "order_id": order_info['order_id'],
        "amount": order_info['amount'],
        "created_at": order_info['created_at']
    }

df = pd.DataFrame([new_random_event() for _ in range(0, 100)])
df
```

## Q1. List all events with "event_type" as "order_created" where "order_fulfilled" is at least 30 days later than the "order_created" (dates are based on "created_at").

```python
#Create new dataframe for order_created and order_fulfilled event type
order_created_df = df.loc[df["event_type"] == "order_created"]
order_fulfilled_df = df.loc[df["event_type"] == "order_fulfilled"]
```

```python

#join the two dataframes to form a new dataframe and save it as df1
df1 = pd.merge(order_created_df, order_fulfilled_df, on = ["order_id", "amount"], how = "inner")
df1.head()
```

```python
#Find the delay in time and storing it as an integer
df1["delayed_days"] = (df1["created_at_y"] - df1["created_at_x"])
df1["delayed_days"] = (df1["delayed_days"])/ np.timedelta64(1, "D")
df1["delayed_days"] = df1["delayed_days"].astype(int)
df1.head()
```

```python
#Remove excess columns
df2 = df1.drop(labels= ["event_id_x", "event_type_x", "event_id_y", "event_type_y"], axis=1)
df2.head()
```

```python
#Rename ambiguous columns
df2.rename(columns= {"created_at_x": "order_date", "created_at_y": "order_fulfilled_date"}, inplace = True)
df2.head()
```

```python
#filter the orders where "order_fulfilled" is at least 30 days later than the "order_created" (i.e, delay_in_time >= 30 days)

df3=df2.loc[df2["delayed_days"] >= 30]
df3.sort_values(by= ["delayed_days"], ascending = False)
```

```python
#converting df_30 to dictionary
df3_dic=df3.to_dict()
df3_dic
```

## Q2. Determine the average, minimum, maximum and count delay per month per calendar year where the delay is greater than 30 days.

```python
#Getting the months and the year from the order date
df3["delayed_month"] = df3['order_date'].dt.month
df3["delayed_year"] = df3['order_date'].dt.year
```

```python
df3
```

```python
#Determine the average(mean) , minimum, maximum of df3 using inbuilt function. That is:
df3.describe()
```

```python
#Grouping the delayed_time metrics by delayed_year and delayed_month
df3_grouped = df3.groupby(["delayed_year", "delayed_month"]).agg(
    {"delayed_days": ["mean", "min", "max", "count"]})
```
